```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(df.year,df.capita,color="blue",marker="*")
```

```python
#Dropping a column
new_df=df.drop('capita',axis='columns')
new_df
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,df.capita)
```

```python
p=model.predict([[2020]])
```
